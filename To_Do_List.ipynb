```
user_input = 'random'
data = []


print('Menu: ')
print('1. Add an Item')
print('2. Mark as done')
print('3. View Items')
print('4. Exit')
print(" ")
name = input("Your name: ")
    

    
while user_input!='4':
    
  
    user_input = input('Enter your choice:\n')
    print(" ")
    
    if user_input=='1':
        item = input("What task do you want to complete?")
        print("--> ")
        data.append(item)
        print('"' + item + '"'+ ' Added')
        print(" ")
        
    elif user_input=='2':
        item = input("What is to be marked as done? ")
        if item in data:
            data.remove(item)
            print("Removed"+" ' " +item+ " ' "+ "from the list")
            print(" ")
        else:
            print("Item does not exist in the list")
            
    elif user_input=='3':
        print("Here's your list: ",name)
        for i in data:
            print(i)
        print(' ')
            
    elif user_input == '4':
        print("Thanks for using the To_Do List." +" "+name+ " " + ", "+" Hope you completed your tasks")
        
    else:
        print("Please enter a number from 1 to 4")
    
```
